# Beginner RNN Basics
M.TECH AI

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

## 1. Generate Data

```python
def generate_data(n=500, seq_len=5):
    X = torch.randn(n, seq_len, 1)
    y = (X.sum(dim=1) > 0).float()
    return X, y

X, y = generate_data()

print(X.shape, y.shape)
```

## 2. Simple RNN Model

```python
class SimpleRNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.rnn = nn.RNN(input_size=1, hidden_size=8, batch_first=True)
        self.fc = nn.Linear(8, 1)

    def forward(self, x):
        h0 = torch.zeros(1, x.size(0), 8)
        out, _ = self.rnn(x, h0)
        out = out[:, -1, :]
        out = self.fc(out)
        return torch.sigmoid(out)

model = SimpleRNN()
```

## 3. Training

```python
loss_fn = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

for epoch in range(50):
    preds = model(X)
    loss = loss_fn(preds, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch+1) % 10 == 0:
        print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
```

## 4. Testing

```python
test_X, test_y = generate_data(5)

with torch.no_grad():
    preds = model(test_X)
    preds = (preds > 0.5).float()

print("Predictions:", preds.squeeze().tolist())
print("Actual:", test_y.squeeze().tolist())
```
