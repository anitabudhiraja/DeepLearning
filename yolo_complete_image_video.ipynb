**YOLO (You Only Look Once)** is a **real-time object detection algorithm** in computer vision that can:

* detect **multiple objects**
* classify them
* and draw **bounding boxes**

—all in a **single forward pass** of a neural network.

---

## 🧠 Core Idea

Traditional detectors (like R-CNN family):

* First find regions → then classify each region (slow)

YOLO:

* **Looks at the entire image once**
* Predicts everything in **one shot**

---

## ⚙️ How YOLO Works (Conceptually)

1. Image is divided into a grid

2. Each grid cell predicts:

   * Bounding boxes
   * Confidence score
   * Class probabilities

3. Final output:

   * Objects + locations + labels

---

## 📦 What YOLO Predicts

For each detected object:

* `(x, y)` → center
* `(w, h)` → width & height
* `confidence` → how sure it is
* `class` → object type (person, car, dog, etc.)

---

## 🚀 Why YOLO is Popular

| Feature            | Benefit                     |
| ------------------ | --------------------------- |
| ⚡ Fast             | Real-time (30–100+ FPS)     |
| 🎯 End-to-end      | No separate region proposal |
| 🧠 Global context  | Sees full image             |
| 🔧 Easy deployment | Works with OpenCV, PyTorch  |

---

## 📊 YOLO Versions (Evolution)

* YOLOv1 → First real-time detector
* YOLOv3 → Widely used (you’re using this)
* YOLOv5 → PyTorch-based (unofficial but popular)
* YOLOv7 → optimized performance
* YOLOv8 → modern, cleaner API, better accuracy

---

## 🔬 In Your Code Context

When you do:

```python
outputs = net.forward(output_layers)
```

YOLO is:

* taking your image
* running a CNN
* outputting **all detections at once**

Then you:

* filter by confidence
* apply **NMS (Non-Max Suppression)**
* draw boxes

---

## 🧩 Real-World Applications

* 🚗 Autonomous driving (cars, pedestrians)
* 📹 CCTV surveillance
* 🏭 Industrial defect detection
* 🛍️ Retail analytics
* 🤖 Robotics vision

---

## ⚠️ Limitations

* Struggles with **very small objects**
* Can miss **overlapping objects**
* Needs **good training data**

# 🧠 Load Model

````
import cv2
import numpy as np
import matplotlib.pyplot as plt

net = cv2.dnn.readNet('yolov3-tiny.weights','yolov3-tiny.cfg')
net.setPreferableBackend(cv2.dnn.DNN_BACKEND_OPENCV)
net.setPreferableTarget(cv2.dnn.DNN_TARGET_CPU)
````

# 🏷️ Load Labels

````
with open('coco.names') as f:
    classes=[line.strip() for line in f]
````

# ⚙️ Output Layers

````
layer_names=net.getLayerNames()
output_layers=[layer_names[i-1] for i in net.getUnconnectedOutLayers()]
````

# 🖼️ Image Inference

**Download Sample Image:** https://www.mediafire.com/file/wvhppk8ihhucwx2/test.jpg/file

````
image=cv2.imread('test.jpg')
h,w=image.shape[:2]
blob=cv2.dnn.blobFromImage(image,1/255.0,(416,416),swapRB=True,crop=False)
net.setInput(blob)
outputs=net.forward(output_layers)

boxes=[];confidences=[];class_ids=[]
for output in outputs:
  for detection in output:
    scores=detection[5:]
    cid=np.argmax(scores)
    conf=scores[cid]
    if conf>0.3:
      cx,cy=int(detection[0]*w),int(detection[1]*h)
      bw,bh=int(detection[2]*w),int(detection[3]*h)
      x,y=int(cx-bw/2),int(cy-bh/2)
      boxes.append([x,y,bw,bh])
      confidences.append(float(conf))
      class_ids.append(cid)

idxs=cv2.dnn.NMSBoxes(boxes,confidences,0.3,0.3)
if len(idxs)>0:
  for i in idxs.flatten():
    x,y,bw,bh=boxes[i]
    label=classes[class_ids[i]]
    cv2.rectangle(image,(x,y),(x+bw,y+bh),(0,255,0),2)
    cv2.putText(image,label,(x,y-10),cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,255,0),2)

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB));plt.axis('off')
````

# 🎥 Video Inference (5 seconds only)

**Download Sample Video:** https://www.mediafire.com/file/3scdo3zhxnyae0m/test.mp4/file

````
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

# Open input video
video = cv2.VideoCapture('test.mp4')

# Get video properties
fps = video.get(cv2.CAP_PROP_FPS)
if fps == 0:
    fps = 30

width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Create output video writer
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('output.mp4', fourcc, fps, (width, height))

# Limit to 5 seconds
max_frames = int(fps * 5)
frame_count = 0

while True:
    success, frame = video.read()

    if not success or frame_count >= max_frames:
        break

    frame_count += 1

    h, w, _ = frame.shape

    # Convert to blob
    blob = cv2.dnn.blobFromImage(frame, 1/255.0, (416, 416),
                                 swapRB=True, crop=False)

    net.setInput(blob)

    # Run model
    detections = net.forward(output_layers)

    boxes = []
    confidences = []
    class_ids = []

    for output in detections:
        for obj in output:
            scores = obj[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]

            if confidence > 0.5:
                cx = int(obj[0] * w)
                cy = int(obj[1] * h)
                bw = int(obj[2] * w)
                bh = int(obj[3] * h)

                x = int(cx - bw / 2)
                y = int(cy - bh / 2)

                boxes.append([x, y, bw, bh])
                confidences.append(float(confidence))
                class_ids.append(class_id)

    indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

    if len(indexes) > 0:
        for i in indexes.flatten():
            x, y, bw, bh = boxes[i]
            label = classes[class_ids[i]]

            cv2.rectangle(frame, (x, y), (x + bw, y + bh),
                          (0, 255, 0), 2)

            cv2.putText(frame, label,
                        (x, y - 10),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.5, (0, 255, 0), 2)

    # Save frame instead of showing
    out.write(frame)

# Release everything
video.release()
out.release()
cv2.destroyAllWindows()

print("Output video saved as output.mp4")
````
